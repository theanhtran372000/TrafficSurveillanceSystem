```python
from icrawler.builtin import GoogleImageCrawler
import os

save_dir = 'data'
amount = 200
google_crawler = GoogleImageCrawler(storage={'root_dir': save_dir})

keywords = [
    'hà nội tắc đường',
    'hải phòng tắc đường',
    'sài gòn tắc đường',
    'đà nẵng tắc đường',
    'việt nam tắc đường',
    'giao thông hà nội',
    'giao thông sài gòn',
    'giao thông hải phòng',
    'giao thông đà nẵng',
    'giao thông việt nam',
    'đường hà nội vắng tanh',
    'đường sài gòn vắng tanh',
    'đường hải phòng vắng tanh',
    'đường đà nẵng vắng tanh',
    'đường việt nam vắng tanh',
    'tắc cao tốc hà nội',
    'tắc cao tốc hải phòng',
    'tắc cao tốc đà nẵng',
    'tắc cao tốc hồ chí minh',
    'tắc cao tốc việt nam',
    'tắc đường vành đai',
    'tắc đường trên cao',
    'hanoi traffic',
    'hanoi traffic jam',
    'hanoi traffic congestion',
    'danang traffic',
    'danang traffic jam',
    'danang traffic congestion',
    'hcm traffic',
    'hcm traffic jam',
    'hcm traffic congestion',
    'hai phong traffic',
    'hai phong traffic jam',
    'hai phong traffic congestion',
    'vietnam traffic'
    'vietnam traffic jam',
    'vietnam traffic congestion',
    'hanoi highway',
    'hanoi freeway',
    'danang highway',
    'danang freeway',
    'hcm highway',
    'hcm freeway',
    'danang highway',
    'danang freeway',
    'vietnam highway',
    'vietnam freeway'
]

for i, kw in enumerate(keywords):
    print('[{}]'.format(i), '= ' * 10)
    print('Crawling for keyword {}'.format(kw))
    
    google_crawler.crawl(
        keyword=kw,
        min_size=(200, 200),
        max_size=None,
        file_idx_offset='auto',
        max_num=amount
    )
    
    print('Done!\n')
```

```python
# Process video
import glob
import cv2

video_dir = 'videos'
start_index = len(os.listdir(save_dir)) + 1
capture_every = 10 # seconds

for video in glob.glob(os.path.join(video_dir, '*', '*', '*')):
    
    print('Process video at {} ...'.format(video))
    cap = cv2.VideoCapture(video)    
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print('Stat: ', fps, frame_count)
    cnt = 0
    
    while True:
        ret, frame = cap.read()
        
        import matplotlib.pyplot as plt
        
        if not ret:
            break
        
        if cnt % (capture_every * fps) == 0:
            
            # Save image
            cv2.imwrite(os.path.join(save_dir, '{}.jpg'.format(str(start_index).zfill(6))), frame)
            start_index += 1
        
        cnt += 1
        
        
```
